```python
import pandas as pd
import re
import os
from sklearn.feature_extraction.text import CountVectorizer

datasets_path = "datasets"
q1_path = "q1"
file_name = "emails_1.csv"

emails_1_path = os.path.join(datasets_path, q1_path, file_name)

df = pd.read_csv(emails_1_path, index_col=False)
df.head()
```

```python
#  Clean text: remove non-letters, lowercase
def clean_text(text:str):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # keep only letters and spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["clean_text"] = df["text"].apply(clean_text)
df["status"] = df["status"].map({"ham": 0, "spam": 1})

df.head()
```

```python
# Bag of Words â€” top 15 frequent words
vectorizer = CountVectorizer(max_features=15)
X_bow = vectorizer.fit_transform(df["clean_text"])

bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())

final_df = pd.concat([df[["status"]], bow_df], axis=1)

print("\nTop 15 words used as features:")
print(vectorizer.get_feature_names_out())

print("\n--- Cleaned and Vectorized Data ---")
final_df.head()
```

```python
import numpy as np

def accuracy(y_true, y_pred):
    """Compute the ratio of correct predictions to total samples."""
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean(y_true == y_pred)

def precision(y_true, y_pred):
    """Compute precision: TP / (TP + FP)."""
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    tp = np.sum((y_pred == 1) & (y_true == 1))
    fp = np.sum((y_pred == 1) & (y_true == 0))
    return tp / (tp + fp) if (tp + fp) != 0 else 0.0

def recall(y_true, y_pred):
    """Compute recall: TP / (TP + FN)."""
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    tp = np.sum((y_pred == 1) & (y_true == 1))
    fn = np.sum((y_pred == 0) & (y_true == 1))
    return tp / (tp + fn) if (tp + fn) != 0 else 0.0

def f1_score(y_true, y_pred):
    """Compute F1 score: harmonic mean of precision and recall."""
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    return 2 * (p * r) / (p + r) if (p + r) != 0 else 0.0


y_true = [0, 1, 1, 0, 1]
y_pred = [0, 1, 0, 0, 1]

print("Accuracy:", round(accuracy(y_true, y_pred), 2))
print("Precision:", round(precision(y_true, y_pred), 2))
print("Recall:", round(recall(y_true, y_pred), 2))
print("F1 Score:", round(f1_score(y_true, y_pred), 2))
```

```python
import numpy as np

class LogisticRegressionScratch:
    def __init__(self, learning_rate=0.01, num_epochs=1000):
        self.learning_rate = learning_rate
        self.num_epochs = num_epochs
        self.weights = None
        self.bias = None

    # Sigmoid Activation
    def sigmoid(self, z):
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))

    # Forward pass
    def compute_output(self, X):
        return np.dot(X, self.weights) + self.bias

    # Binary Cross Entropy Loss
    def compute_loss(self, y, y_pred):
        m = len(y)
        # Add small epsilon to avoid log(0)
        epsilon = 1e-9
        loss = - (1 / m) * np.sum(
            y * np.log(y_pred + epsilon) + (1 - y) * np.log(1 - y_pred + epsilon)
        )
        return loss

    def compute_gradients(self, X, y, y_pred):
        m = X.shape[0]
        dw = (1 / m) * np.dot(X.T, (y_pred - y))
        db = (1 / m) * np.sum(y_pred - y)
        return dw, db

    def update_weights(self, dw, db):
        self.weights -= self.learning_rate * dw
        self.bias -= self.learning_rate * db

    # Training
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.random.randn(n_features) * 0.01
        self.bias = 0

        for epoch in range(self.num_epochs):
            linear_output = self.compute_output(X)

            y_pred = self.sigmoid(linear_output)

            dw, db = self.compute_gradients(X, y, y_pred)

            self.update_weights(dw, db)

            if epoch % 10 == 0:
                loss = self.compute_loss(y, y_pred)
                print(f"Epoch {epoch} â€” Loss: {loss:.4f}")

    def predict(self, X):
        linear_output = self.compute_output(X)
        y_pred_prob = self.sigmoid(linear_output)
        return np.where(y_pred_prob >= 0.5, 1, 0)
```

```python
import numpy as np

class MultinomialNaiveBayes:
    def __init__(self, alpha=1.0):
        self.alpha = alpha # Laplace smoothing parameter
        self.class_log_prior_ = None  # log prior probabilities of classes
        self.feature_log_prob_ = None  # log probability of features per class
        self.classes_ = None  # unique class labels

    def fit(self, X, y):
        # X: feature matrix (n_samples, n_features)
        # y: target labels (n_samples,)
        n_samples, n_features = X.shape
        self.classes_ = np.unique(y)
        n_classes = len(self.classes_)

        # Compute log prior probability for each class
        self.class_log_prior_ = np.zeros(n_classes)
        for idx, c in enumerate(self.classes_):
            X_c = X[y == c]
            self.class_log_prior_[idx] = np.log(X_c.shape[0] / n_samples)

        # Compute log probability of features per class with Laplace smoothing
        self.feature_log_prob_ = np.zeros((n_classes, n_features))
        for idx, c in enumerate(self.classes_):
            X_c = X[y == c]
            # sum of feature counts + alpha for smoothing
            smoothed_fc = X_c.sum(axis=0) + self.alpha
            smoothed_total = smoothed_fc.sum()
            # log probabilities of features
            self.feature_log_prob_[idx, :] = np.log(smoothed_fc / smoothed_total)

    def predict_log_proba(self, X):
        #Compute log-probabilities of each class for input samples.
        return X @ self.feature_log_prob_.T + self.class_log_prior_

    def predict(self, X):
        log_probs = self.predict_log_proba(X)
        return self.classes_[np.argmax(log_probs, axis=1)]
```
